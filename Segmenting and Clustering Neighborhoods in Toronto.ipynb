```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
len(df)
```

```python
df[0]
```

```python
#make the first wikipedia table as the target dataframe we want to work ont
toronto_df=df[0]
toronto_df
```

```python
#make a copy for back up
toronto_df_copy=toronto_df.copy()
toronto_df_copy
```

```python
#drop cells that with a borough that is Not assigned
indexNames=toronto_df[(toronto_df['Borough']=='Not assigned')].index
indexNames
```

```python
toronto_df1=toronto_df.drop(indexNames, inplace=True)
```

```python
#renumbering the index
toronto_df1=toronto_df.reset_index(drop=True)
toronto_df1
```

```python
#number of rows in the dataframe
toronto_df1.shape
```

## Importing the csv file containing the latitudes & longitudes in Canada

```python
geodata=pd.read_csv('http://cocl.us/Geospatial_data')

geodata.head()
```

## Merge the two tables 

```python
toronto_geo_data=pd.merge(toronto_df1, geodata, on = 'Postal Code')
toronto_geo_data.head()
```

## make a new table that contain "Toronto" in Borough column

```python
toronto_final_df=toronto_geo_data[toronto_geo_data['Borough'].str.contains('Toronto',regex=False)]
toronto_final_df
```

```python
toronto_final_df.shape
```

## Visualize all the chosen neighborhood 

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

for lat,lng,borough,neighbourhood in zip(toronto_final_df['Latitude'],toronto_final_df['Longitude'],toronto_final_df['Borough'],toronto_final_df['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

### Devided the area into clusters using Kmeans clustering

```python
k=5
toronto_clustering = toronto_final_df.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
toronto_final_df.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
toronto_final_df
```

### Visualization of the clusters

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(toronto_final_df['Latitude'], toronto_final_df['Longitude'], toronto_final_df['Neighbourhood'], toronto_final_df['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
